1.Create a sequence of a given length from a given start point, where the difference between 2 consecutive elements of the expected sequence is also given as step.

```python
import numpy as np
```

```python
def seq(start,length,step):
  a=start+(length-1)*step # Finding endpoint from the start, length and step given
  arr=np.linspace(start,a,length,dtype='int32') 
  return arr
seq(5,10,3)
```

5. Given a NumPy matrix of shape (n, n), remove all the odd elements from the matrix and return a new matrix of shape (k, k) consisting of all the remaining even elements from the original matrix.

10.Given a NumPy array of integer elements, write a program that calculates the sigmoid value of each of those elements and returns these sigmoid values rounded up to 2 decimal places in a NumPy array.

The value of sigmoid function for an element x is:

Sigmoid(x) =1/(1+e^โx)

```python
def sigmoid(arr):
  arr=np.array(arr)
  arr=1/(np.exp(arr*-1)+1) #Calculate formula
  return np.round(arr,2) #Round to 2 decimals
sigmoid([-2,-1,0,1,2])
```

12.Given a NumPy array, return the index of the first element that is greater than the target passed as an argument.

Return -1, if there isn't any element greater than the target
